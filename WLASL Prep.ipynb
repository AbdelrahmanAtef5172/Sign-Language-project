# WLASL Preprocessing Pipeline

This notebook handles:
- Loading WLASL dataset
- Extracting metadata (gloss, video_id, URLs)
- Converting raw JSON into structured DataFrame
- Saving cleaned data for further processing (ML pipeline)

```python
import numpy as np 
import pandas as pd 
import os
```

```python
for root, dirs, files in os.walk('/kaggle/input'):
    for file in files:
        if file.endswith('.json'):
            print(os.path.join(root, file))
```

```python
import json
with open('/kaggle/input/datasets/risangbaskoro/wlasl-processed/WLASL_v0.3.json') as f: 
    data = json.load(f)
print("Data Loaded")
```

```python
print(type(data))
print("Number of words:", len(data))
print("Sample:\n", data[0])
```

```python
rows = []

for item in data:
    gloss = item['gloss']
    
    for inst in item['instances']:
        rows.append({
            "video_id": inst.get("video_id"),
            "gloss": gloss,
            "url": inst.get("url"),
        })

df = pd.DataFrame(rows)

print("DataFrame Created ")
print(df.head())
print("Shape:", df.shape)
```

```python
print("Unique words:", df['gloss'].nunique())
print("Total videos:", len(df))
```

```python
os.makedirs('/kaggle/working/metadata', exist_ok=True)

df.to_csv('/kaggle/working/metadata/dataframe.csv', index=False)

print("DataFrame saved")
```

```python
urls_df = df[['url']]

urls_df.to_csv('/kaggle/working/metadata/urls.csv', index=False)

print("URLs saved")
```
