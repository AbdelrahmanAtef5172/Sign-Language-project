```python
import os
import shutil

os.makedirs("voices", exist_ok=True)

voice_files = [
    "female_neutral.wav",
    "female_happy.wav",
    "female_sad.wav",
    "female_angry.wav",
    "male_neutral.wav",
    "male_happy.wav",
    "male_sad.wav",
    "male_angry.wav",
]

for file in voice_files:
    if os.path.exists(file):
        shutil.move(file, f"voices/{file}")
    else:
        print(f"⚠️ Missing file: {file}")
```

```python
os.listdir("voices")
```

```python
import torch
from TTS.api import TTS

# detect GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

# load model
tts = TTS(model_name="tts_models/multilingual/multi-dataset/xtts_v2").to(device)

voice_map = {
    ("female", "neutral"): "voices/female_neutral.wav",
    ("female", "happy"):   "voices/female_happy.wav",
    ("female", "sad"):     "voices/female_sad.wav",
    ("female", "angry"):   "voices/female_angry.wav",

    ("male", "neutral"):   "voices/male_neutral.wav",
    ("male", "happy"):     "voices/male_happy.wav",
    ("male", "sad"):       "voices/male_sad.wav",
    ("male", "angry"):     "voices/male_angry.wav",
}

def generate_speech(text, gender, emotion, output_path="output.wav"):
    speaker_wav = voice_map.get((gender, emotion))

    if speaker_wav is None:
        raise ValueError(f"No voice found for {gender} + {emotion}")

    if not os.path.exists(speaker_wav):
        raise FileNotFoundError(f"Voice file not found: {speaker_wav}")

    tts.tts_to_file(
        text=text,
        speaker_wav=speaker_wav,
        language="en",
        file_path=output_path
    )

    print(f"✅ Speech saved to {output_path}")
```

```python
generate_speech(
    text="I lost my job today",
    gender="female",
    emotion="sad"
)
```
